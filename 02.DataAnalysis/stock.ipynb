```python
seoulPark = pd.read_csv('./static/data/서울시 공원.csv')
```

```python
seoulPark
```

```python
import folium
```

```python
map = folium.Map(location=[seoulPark[seoulPark['구별']=='양천구']['lat'].mean(),
seoulPark[seoulPark['구별']=='양천구']['lng'].mean()], zoom_start=12)
for i in seoulPark[seoulPark['구별']=='양천구'].index:
    folium.Marker([seoulPark.lat[i], seoulPark.lng[i]], popup=seoulPark['공원명'][i],
                                tooltip=seoulPark['면적'][i]).add_to(map)
map
```
